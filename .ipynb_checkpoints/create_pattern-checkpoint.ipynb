```python
import re
import json
```

```python
with open('pattern.svg', 'r') as f:
    lines = f.readlines()
    
lines
```

```python
def parse_rect(rect_str):
    floss = {
        'st0': {'code':'703', 'symbol':'∩', 'color':'#66A647'},
        'st1': {'code':'702', 'symbol':'/', 'color':'#138D46'},
        'st2': {'code':'700', 'symbol':'+', 'color':'#007443'},   
        'st3': {'code':'973', 'symbol':'>', 'color':'#FFC617'},
        'st4': {'code':'743', 'symbol':'⨯', 'color':'#FBAD31'},
        'st5':  {'code':'445', 'symbol':'S', 'color':'#EDE66A'},
        'st6': {'code':'562', 'symbol':'H', 'color':'#1E8061'},
        'st7': {'code':'561', 'symbol':'Z', 'color':'#386654'},
        'st8': {'code':'15', 'symbol':'-', 'color':'#C0DC94'},
        'st9': {'code':'352', 'symbol':'→', 'color':'#F26E66'},
        'st10': {'code':'606', 'symbol':'∪', 'color':'#EE282B'},
        'st11': {'code':'608', 'symbol':'|', 'color':'#EF472A'},
        'st12': {'code':'351', 'symbol':'<', 'color':'#EC404E'},
        'st13': {'code':'761', 'symbol':'Ɪ', 'color':'#F69FA5'},
        'st14': {'code':'Blanc', 'symbol': '○', 'color':'#FDFDFD'},
        'st15': {'code':'321', 'symbol':'↑', 'color':'#BF2148'},
        'st16': {'code':'563', 'symbol':'\\', 'color':'#67B68E'}
    }
    
    x_pattern = 'x="(\d*\.?\d*)"' 
    y_pattern = 'y="(\d*\.?\d*)"'
    class_pattern = 'class="(st\d*)"'
    x = int((float(re.search(x_pattern, rect_str).group(1)) - 243.16)/5.17) + 8
    y = int((float(re.search(y_pattern, rect_str).group(1)) - 295.4)/5.17) + 20
    color = re.search(class_pattern, rect_str).group(1)
    return {'x':x, 'y':y, 'color':floss[color]}
```

```python
squares = [parse_rect(line) for line in lines if 'rect' in line]
squares
```

```python
with open('pattern.json', 'w', encoding='utf-8') as f:
    json.dump(squares, f)
```
